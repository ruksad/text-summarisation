# **Overview of Text Summarization Techniques in the Notebook**

This notebook showcases two distinct approaches to text summarization:  
1. **Extractive Summarization using spaCy**  
2. **Abstractive Summarization using T5**

---

## **1. Extractive Summarization (spaCy)**

The extractive summarization workflow involves the following steps:

- The input text is processed using **spaCy**, creating a document object for linguistic analysis.
- Tokens are extracted and cleaned by:
  - Removing stop words and punctuation  
  - Converting remaining tokens to lowercase  
- Significant words are counted to compute their **frequency**, which is then normalized to produce **importance scores**.
- The text is split into **individual sentences**, and each sentence is assigned a score based on the cumulative importance scores of its words.
- Using `heapq.nlargest`, the highest-scoring sentences are selected to form the **extractive summary**.

An interactive **ipywidget** is included, allowing users to:

- Enter custom input text  
- Specify the desired number of sentences  
- Generate extractive summaries dynamically using the spaCy-based method  

---

## **2. Abstractive Summarization (T5)**

A second approach uses **Hugging Face Transformers** to perform **abstractive summarization**, which generates new, concise sentences that may not appear in the original text.

Key steps include:

- Initializing a pre-trained **`t5-base`** model within a Transformers summarization pipeline  
- Applying the pipeline to generate coherent and context-aware summaries that reinterpret the input text  

A dedicated **ipywidget** is also provided, enabling users to:

- Input text interactively  
- Apply the T5-based abstractive summarization model  
- Compare results with the extractive method  

---

```python
import ipywidgets as widgets
from IPython.display import display
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from collections import Counter
from heapq import nlargest

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

def summarize_text_spacy(text, num_sentences):
    doc = nlp(text)
    tokens = [token.text.lower() for token in doc
              if not token.is_stop and not token.is_punct and token.text != '\n']

    word_freq = Counter(tokens)
    if not word_freq:
        return "No words found in the text."

    max_freq = max(word_freq.values())
    for word in word_freq.keys():
        word_freq[word] = word_freq[word]/max_freq

    sent_token = [sent.text for sent in doc.sents]

    sent_score = {}
    for sent in sent_token:
        for word in sent.split():
            if word.lower() in word_freq.keys():
                if sent not in sent_score.keys():
                    sent_score[sent] = word_freq[word.lower()] # Use word.lower() here to match the keys in word_freq
                else:
                    sent_score[sent] += word_freq[word.lower()] # Use word.lower() here as well

    if not sent_score:
        return "Could not score any sentences for summarization."

    # Select top-scoring sentences
    # Ensure num_sentences does not exceed the number of available sentences
    actual_num_sentences = min(num_sentences, len(sent_score))
    summarized_sentences = nlargest(actual_num_sentences, sent_score, key=sent_score.get)

    return " ".join(summarized_sentences)

# Create ipywidgets
text_input = widgets.Textarea(
    value='',
    placeholder='Enter text to summarize',
    description='Text:',
    disabled=False,
    layout=widgets.Layout(width='80%', height='150px')
)

num_sentences_input = widgets.IntText(
    value=3,
    description='Number of Sentences:',
    disabled=False
)

summarize_button = widgets.Button(
    description='Summarize',
    disabled=False,
    button_style='info', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to summarize',
    icon='file-alt' # (FontAwesome icons)
)

output_area = widgets.Output()

def on_summarize_button_clicked(b):
    with output_area:
        output_area.clear_output()
        text_to_summarize = text_input.value
        num_sents = num_sentences_input.value
        if text_to_summarize:
            summary = summarize_text_spacy(text_to_summarize, num_sents)
            print("Summary:")
            print(summary)
        else:
            print("Please enter some text to summarize using extraction")


summarize_button.on_click(on_summarize_button_clicked)

# Arrange widgets
display(text_input, num_sentences_input, summarize_button, output_area)
```
